```python

# Set Parameters for API
weather_url = "http://api.worldweatheronline.com/premium/v1/weather.ashx?"
weather_response = requests.get(weather_url,params={
                                                  'q': 'Washington',
                                                  'date': '2019-05-01',
                                                  'mca':'yes',
                                                  'format':'json',
                                                  'key': weather_api})
        
   
```

```python
weather_response
```

```python
weather_result = weather_response.json()
df_weather = pd.json_normalize(weather_result)
df_weather
```

```python
month_2019 = weather_result['data']['ClimateAverages'][0]['month']
```

```python
df_weather = pd.json_normalize(month_2019, errors='ignore')
```

```python
df = pd.DataFrame(df_weather)
df.rename(columns={"index": "Month"})
```
