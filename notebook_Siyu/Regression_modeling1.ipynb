```python
for x in numerical:
    df[x] = scaler.fit_transform(np.array(df[x]).reshape(-1, 1))
```

```python
df.head()
```

```python
X = df.drop(columns=['target'])
y = scaler.fit_transform(np.array(df['target']).reshape(-1, 1))
```

```python
from sklearn.decomposition import PCA 

pca = PCA(n_components=800)

X_pca = pca.fit_transform(X)
```

```python
X_pca.shape
```

#### Regession model testing


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, train_size=0.8, random_state=0)
```

XGB

```python
from xgboost import XGBRegressor

xgbr = XGBRegressor(eta=0.05, gamma=1, 
                max_depth=4, subsample=0.8,
                alpha=0.9)


xgbr.fit(X_train,y_train)
y_pred = xgbr.predict(X_test)

print(f"train_r2 = {xgbr.score(X_train,y_train)}")
print(f"test_r2 = {r2_score(y_test,y_pred)}")
print(f"MSE = {mean_squared_error(y_test, y_pred)}")
print(f"RMSE = {np.sqrt(mean_squared_error(y_test, y_pred))}")
print(f"MAE = {mean_absolute_error(y_test, y_pred)}")

```
